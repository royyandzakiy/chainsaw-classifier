```python
import scipy
from scipy import signal
from scipy.io import wavfile

import matplotlib.pyplot as plt
import librosa
import os
# import pyaudio
import wave
import sys
import matplotlib.image as mpimg
import IPython.display as ipd
import numpy as np

directory = 'G:\\Downloads\\ESC-50-master\\ESC-50-master\\audio'
train_X = []
train_y = []

CHUNK = 1024
pre_emphasis = 0.97

def main():
    count = 0
    for file in os.listdir(directory):
        filename = file
        count+=1
        print(str(count) + ") " + filename)
                
        if filename.endswith("-41.wav"):
            filepath = os.path.join(directory, filename)
                            
            sample_rate, samples = wavfile.read(filepath)
            target_sample_rate = (16000/sample_rate)*len(samples) # ini rumus dr mana?
            samples = signal.resample(samples, int(target_sample_rate))

            # normal sound
            frequencies, times, spectrogram = signal.spectrogram(samples, target_sample_rate)
            
            print("plot normal sound:")
            plt.pcolormesh(times, frequencies, spectrogram)
            plt.ylabel('Frequency [Hz]')
            plt.xlabel('Time [sec]')
            plt.show()
            
            # emphasized_signal
            samples = np.append(samples[0], samples[1:] - pre_emphasis * samples[:-1])
            
            frequencies, times, spectrogram = signal.spectrogram(samples, target_sample_rate)
            
            print("plot emphasized_signal sound:")
            plt.pcolormesh(times, frequencies, spectrogram)
            plt.ylabel('Frequency [Hz]')
            plt.xlabel('Time [sec]')
            plt.show()
            
            # fft sound
#             samples_fft = abs(np.fft.lfft(samples)), abs(np.fft.rfft(samples))
#             plt.figure(1)
#             a = plt.subplot(211)
#             r = 2**16/2
#             a.set_ylim([-r, r])
#             a.set_xlabel('time [s]')
#             a.set_ylabel('sample value [-]')
#             x = np.arange(len(samples_fft))/len(samples_fft)
#             plt.plot(x, samples_fft)
            
#             b = plt.subplot(212)
#             b.set_xscale('log')
#             b.set_xlabel('frequency [Hz]')
#             b.set_ylabel('|amplitude|')
#             plt.plot(samples_fft)
#             plt.savefig('sample-graph.png')

            # fft sound 2
#             X = scipy.fft(samples)
#             X_mag = np.absolute(X)
#             f = np.linspace(0, sample_rate, len(X_mag)) # frequency variable
            
#             print("plot magnitude spectrum:")
#             plt.figure(figsize=(13, 5))
#             plt.plot(f, X_mag) # magnitude spectrum
#             plt.xlabel('Frequency (Hz)')
            
#             print("plot magnitude spectrum (zoomed in):")
#             plt.figure(figsize=(13, 5))
#             plt.plot(f[:5000], X_mag[:5000])
#             plt.xlabel('Frequency (Hz)')
            
            # mfcc sound
            mfccs = librosa.feature.mfcc(samples, sr=sample_rate)
            print("mfcc shape:")
            print(mfccs.shape)
            
            print("print mfcc:")
            print(mfccs)
            
            print("samples shape:")
            print(samples.shape)
            
            print("sample_rate shape:")
            print(sample_rate)
            
            print("plot mfcc coefficient index:")
            plt.imshow(mfccs.T, origin='lower', aspect='auto', interpolation='nearest')
            plt.ylabel('MFCC Coefficient Index')
            plt.xlabel('Frame Index')

            # mpimg.imsave("spectograms\\spectogram-" + filename[:-4], spectrogram) # simpan spektorgram dalam file

#             print("play audio:")
#             ipd.display(ipd.Audio(filepath))

            break

if __name__ == "__main__":
    main()
```
